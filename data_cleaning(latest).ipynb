```python
def preprocess(text):
    stop_words = set(stopwords.words('english'))
    filtered_words=[]

    list_text=re.findall(r'\w+(?:\.\w+)?', text)
    ans=[]
    for i in range(len(list_text)):
      if list_text[i].lower() not in stop_words:
        filtered_words.append(list_text[i])
    
    return filtered_words
```
